```r
library('tidyverse')
bodyfat = read_csv('bodyfat.csv')
colnames(bodyfat)
B_F = bodyfat[order(bodyfat$bodyfat_percentage,decreasing=TRUE),] # sorting to decreasing value
colnames(B_F)
```

```r
# create a categorical variable (with multiple possible values)
B_F$BMI = as.factor(ifelse(B_F$bodyfat_percentage > 18, 'over', 'normal'))
# view updated data frame
B_F
```

```r
install.packages("devtools")
library("devtools")
install_github("kassambara/factoextra")
```

```r
library(factoextra)
library(ggfortify)
```

```r
# 1. carrying out PCA on BF(bodyfat dataset)
# extract the seven body circumference measurements
BF_sub = select(B_F, -id, -bodyfat_percentage, -age, -weight, -height, -BMI)
```

```r
# a. compute PCA for **_sub
BF_PCA = prcomp(BF_sub, scale.=TRUE)
```

```r
# scree plot - ref: scree: Draws a scree plot.In BinMat: Processes Binary Data Obtained from Fragment Analysis (Such as AFLPs, ISSRs, and RFLPs) https://rdrr.io/cran/BinMat/man/scree.html
fviz_eig(BF_PCA, addlabels=TRUE, geom="line")
```

Interpretation:
with the PCA for the seven body measurement in plot, PC1 is responsible for more than 50% of the explained variables with PC2 in que (TDH, 2022). Both summed up make up a strong relation and eigenvalue line of 74%. PC3 and PC4 forms another straight line with PC2, therefore, accounting for 40% of the principle components.

Reference:
Train Data Hub; May 25, 2022; “How to Create A scree Plot in R and How to Interpret Them”; https://traindatahub.com/?p=297


```r
# biplot
autoplot(BF_PCA, data=bodyfat, colour = 'bodyfat_percentage',
         label=TRUE, label.size=3, shape=FALSE, loadings.colour = 'magenta',
         loadings=TRUE, loadings.label=TRUE)
```

Interpretation:
*   In this context, PC1 explains 52.69% of variablity and PC2 explains 17.94% so together they explain 74% of variability of the data set.
*   PC1 is positively correlated with the forearm on a single entity, while the neck, wrist and biceps are closely in range of each other.
*   The ankle, hip and knee are closely correlated which forms another strong bond under a percentage of the bodyfat_percentage.
*   Following the direction of the points/labels, it could be said that those pointed in the increased direction from the center point, are heavily affected by bodyfat, and the more the percentage, the heavier theses parts are strained.
*   However, the downward decreasing labels can be said to not have a heavier effect of the variable under consideration, while the forearm has no bearing to the presentation.








```r
# loadings plot
loadings = as.data.frame(BF_PCA$rotation[,1:7])
loadings$Symbol = row.names(loadings)
loadings = gather(loadings, key='Component', value='Weight', -Symbol)
ggplot(loadings, aes(x=Symbol,y=Weight)) +
  geom_bar(stat='identity') +
  facet_grid(Component~.)
variance_new = (BF_PCA$sdev)^2
variance_explained = variance_new/sum(variance_new)
variance_explained
```

Interpretation:
*   PC1, PC2 and PC3 explains about 85% of the variability which is good.
*   PC1 is strongly correlated at the hip and knee and absolutely or narrowly disengaged from the forearm.
*   While the other PCs also do not show strong correlations with the forearm, PC3 hold the vast amount of correlated to the variable. It also shows in the direction of the plot, weight is not a factor for its representation.
*   PC7 has an equal strong inluence on the increasing and decreasing weight for the wrist, so it is safe to say this part of the body if the bodyfat percentage is over bound or normal, the wrist will have either implications.
*   The biceps on the other hand is only heavily correlated at PC5 under less weight evaluation.






```r
# biplot for PCA2 and PCA3 only
autoplot(BF_PCA, data = B_F, x = 2, y = 3, colour = 'BMI',
         label=TRUE, label.size=3, loadings.colour = 'purple',
         loadings=TRUE, loadings.label=TRUE) # ref: autoplot.pca_common: Autoplot PCA-likes in ggfortify: Data Visualization Tools for Statistical Analysis Results
```

Interpretation:
*  There is a strong correlation in PC2 with the hip, knee and ankle with positive loadings. Thus, we maintain, that the categorical variable (BMI) of over(in blue) is a determinant for the behaviour of these parts.
*  The neck, wrist and biceps though strongly correlated are in between readings, as affected by either the over weight or normal range.
*  The forearm once again though represented by PC2, is still solo and not necessarily influenced by either over or normal ranges. Thus, there is a weak correlation with this variable.

```r
# loadings plot
loadings = as.data.frame(BF_PCA2$rotation[,1:10])
loadings$Symbol = row.names(loadings)
loadings = gather(loadings, key='Component', value='Weight', -Symbol)
ggplot(loadings, aes(x=Symbol,y=Weight)) +
  geom_bar(stat='identity') +
  facet_grid(Component~.)
variance_new = (BF_PCA2$sdev)^2
variance_explained = variance_new/sum(variance_new)
variance_explained
```

interpretation:

*   Though PC1 axis has strong correlated with most of the variables in this plot, and it carries 47% of the explained variability(good), PC10, PC3, PC4 and PC5 also show strong correlations of various body parts in context.
*   In most cases, forearm and height seem not to have any distinct correlation with any principal component and in most cases like in PC8, height is completely non-linked to any feature.
*   Weight also, has a severed relationship in PC4 or better yet, non-existent.

```r
# 2 Cluster Analysis
library(cluster)
# 2(a)people(rows)
BF1=B_F[6:12]
```

```r
D_Euc=dist(scale(BF1), method="euclidean")
cluster_results=agnes(D_Euc, method="ward")
plot(cluster_results, which.plots=2, labels=B_F$BMI)
rect.hclust(cluster_results,k=4)
```

```r
D_ManH=dist(scale(BF1), method="manhattan")
cluster_results=agnes(D_ManH,method="ward")
plot(cluster_results, which.plots=2, labels=B_F$BMI)
rect.hclust(cluster_results,k=3)
```

```r
D_CanB=dist(scale(BF1), method="canberra")
cluster_results=agnes(D_CanB,method="gaverage")
plot(cluster_results, which.plots=2, labels=B_F$BMI)
rect.hclust(cluster_results,k=3)
```

```r
D_Mink=dist(scale(BF1),method="minkowski")
cluster_results=agnes(D_Mink, method="weighted")
plot(cluster_results, which.plots=2, labels=B_F$BMI)
rect.hclust(cluster_results, k=2)
```

```r
# 2b
D=dist(t(scale(BF1)),method="maximum")
cluster_results=agnes(D,method="ward")
plot(cluster_results,which.plots=2)
rect.hclust(cluster_results,k=3)
```

Inferences:
The dendogram consist of Three clusters
The left most cluster consist of neck,waist and biceps
The middle cluster consist of hip,knee and ankle
The right most cluster consist of forearm
The Forearm is treated as an Outlier
The Distance metric used here is "maximum"i.e It measures the distances of the maximum difference between two points over any of their axis values.  
The agnes method used here is "ward".
The ideal number of cluster taken must be between 2 and 6.
If we take 2 clusters the forearm will be at Right cluster and The rest of the measurement will be at left clusters these does not seems to be an good fit
if we take 4,5 and 6 clusters there will be some inappropriate clustering.So,the ideal number of clusters taken is 3.

```r
# 2(b.ii) 10 variables
BF2=B_F[3:12]
D=dist(t(scale(BF2)),method="maximum")
cluster_results=agnes(D,method="ward")
plot(cluster_results,which.plots=2)
rect.hclust(cluster_results,k=3)
```
