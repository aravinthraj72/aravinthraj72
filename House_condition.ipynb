```python
X3=X1[['grade', 'sqft_lot', 'waterfront', 'Year', 'Day', 'id', 'Weekday',
       'lat', 'yr_renovated', 'bedrooms', 'sqft_lot15', 'view', 'long']]
y3=y1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X3,y3,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score,auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
# Define the parameter values
max_depth_values = [250,500,400,300]
min_samples_split_values = [500,400,20,10,5,2]

# Initialize variables to store the best accuracy and corresponding parameters
best_accuracy = 0
best_params = {}

# Initialize lists to store the FPR and TPR for each model
fpr_list = []
tpr_list = []
auc_list = []

# Binarize the labels for multi-class classification
y_train_bin = label_binarize(y_train, classes=np.unique(y_train))
y_test_bin = label_binarize(y_test, classes=np.unique(y_train))

# Iterate over all parameter combinations
for max_depth in max_depth_values:
    for min_samples_split in min_samples_split_values:
        # Create a Decision Tree classifier with the current parameter values
        clf = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split)

        # Fit the classifier to the training data
        clf.fit(X_train, y_train)

        # Predict the probabilities for each class
        y_scores = clf.predict_proba(X_test)

        # Compute the ROC curve and AUC for each class
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(len(np.unique(y_train))):
            fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_scores[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        # Append the FPR, TPR, and AUC to the respective lists
        fpr_list.append(fpr)
        tpr_list.append(tpr)
        auc_list.append(roc_auc)

        # Calculate the accuracy for the current parameter combination
        accuracy = clf.score(X_test, y_test)

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'max_depth': max_depth, 'min_samples_split': min_samples_split}

# Plot the ROC curves for all models
plt.figure(figsize=(8, 6))
colors = cycle(['blue', 'red', 'green','yellow','orange'])  # You can adjust the colors as needed
for i, color in zip(range(len(np.unique(y_train))), colors):
    for fpr, tpr, auc, max_depth, min_samples_split in zip(fpr_list, tpr_list, auc_list, max_depth_values, min_samples_split_values):
        label = f"Class {i+1} (AUC = {auc[i]:.2f})"
        plt.plot(fpr[i], tpr[i], color=color, label=label)

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line representing random classifier
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve (OvR)')
plt.legend(loc='lower right',fontsize='small')
plt.show()

# Print the best parameters and accuracy
print("Best Parameters:", best_params)
print("Best Accuracy:", best_accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle

# Define the parameter values
n_neighbors_values = [5, 7, 10]
weights_values = ['uniform', 'distance']

# Initialize variables to store the best accuracy and corresponding parameters
best_accuracy = 0
best_params = {}

# Initialize lists to store the FPR and TPR for each model
fpr_list = []
tpr_list = []
auc_list = []

# Binarize the labels for multi-class classification
y_train_bin = label_binarize(y_train, classes=np.unique(y_train))
y_test_bin = label_binarize(y_test, classes=np.unique(y_train))

# Iterate over all parameter combinations
for n_neighbors in n_neighbors_values:
    for weights in weights_values:
        # Create a KNN classifier with the current parameter values
        clf = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)

        # Fit the classifier to the training data
        clf.fit(X_train, y_train)

        # Predict the probabilities for each class
        y_scores = clf.predict_proba(X_test)

        # Compute the ROC curve and AUC for each class
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        for i in range(len(np.unique(y_train))):
            fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_scores[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        # Append the FPR, TPR, and AUC to the respective lists
        fpr_list.append(fpr)
        tpr_list.append(tpr)
        auc_list.append(roc_auc)

        # Calculate the accuracy for the current parameter combination
        accuracy = clf.score(X_test, y_test)

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'n_neighbors': n_neighbors, 'weights': weights}

# Plot the ROC curves for all models
plt.figure(figsize=(8, 6))
colors = cycle(['blue', 'red', 'green', 'yellow', 'orange'])  # You can adjust the colors as needed
for i, color in zip(range(len(np.unique(y_train))), colors):
    for fpr, tpr, auc, n_neighbors, weights in zip(fpr_list, tpr_list, auc_list, n_neighbors_values, weights_values):
        label = f"Class {i+1} (AUC = {auc[i]:.2f})"
        plt.plot(fpr[i], tpr[i], color=color, label=label)

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line representing random classifier
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve (OvR)')
plt.legend(loc='lower right', fontsize='small')
plt.show()

# Print the best parameters and accuracy
print("Best Parameters:", best_params)
print("Best Accuracy:", best_accuracy)
```
